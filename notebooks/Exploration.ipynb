```
import pandas as pd
from google.cloud import bigquery

class Big_Query_Helper:
  def __init__(self, data_project, dataset=None):
    self.client = bigquery.Client(data_project)
    if dataset:
      self.data_ref = self.client.dataset(dataset)
      self.dataset = self.client.get_dataset(self.data_ref)
    else:
      self.data_ref = None
      self.dataset = None

  def list_table_names(self):
    if not self.dataset:
      raise Exception('Dataset not defined')

    self.tables = self.client.list_tables(self.dataset)

    return [_.table_id for _ in self.tables]

  def preview_table(self, table_name, projection=None, max_results=20):
    if not self.dataset:
      raise Exception('Dataset not defined')
    tab = self.client.get_table(self.dataset.table(table_name))
    results = self.client.list_rows(tab, selected_fields=projection,
                                    max_results=max_results, start_index=None)
    results = [x for x in results]
    return pd.DataFrame(data=[list(x_.values()) for x_ in results],
                        columns=list(results[0].keys()))
```

```
# Create Reference and fetch the Stack Overflow dataset
so_query_helper = Big_Query_Helper(data_project='bigquery-public-data',
                                   dataset='stackoverflow')

# Show the list of Stack Overflow tables
so_tables = so_query_helper.list_table_names()
print('Stack Overflow Dataset has the following tables : \n{}'.format('\n'.join(
    '{}' for _ in range(len(so_tables))).format(*so_tables)))
```
